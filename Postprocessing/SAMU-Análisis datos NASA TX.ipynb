# üöÄ Interpretaci√≥n del NASA-TLX (NASA-Task Load Index)

El NASA-TLX es una herramienta **comparativa** para evaluar la **carga mental subjetiva** de una tarea, con una
puntuaci√≥n final en una escala de **0 a 100**.

---

## üî¢ Interpretaci√≥n del √çndice Global (Puntuaci√≥n Final)

La puntuaci√≥n global indica el nivel general de demanda de la tarea.

| Rango de Puntuaci√≥n (0-100) | Interpretaci√≥n General |
| :--- | :--- |
| **0 - ~20** | **Carga Muy Baja:** Tarea muy f√°cil y con poco esfuerzo. |
| **~40 - ~60** | **Carga Media/Moderada:** Demanda notable, pero sostenible. |
| **~80 - 100** | **Carga Muy Alta/Excesiva:** Tarea llevada al l√≠mite, alto riesgo de fatiga y error. |

> **Uso Principal:** El valor es m√°s √∫til cuando se **compara** una condici√≥n (ej. interfaz A) con otra (ej. interfaz B) para determinar cu√°l es menos demandante.

---

## üìä Interpretaci√≥n de las 6 Dimensiones Individuales

Observar las puntuaciones altas en las subescalas permite identificar **la naturaleza** de la carga (d√≥nde se concentra la demanda).

| Dimensi√≥n | Significado de Puntuaci√≥n Alta | Implicaci√≥n |
| :--- | :--- | :--- |
| **Exigencia Mental** | Alta necesidad de pensar, decidir o recordar. | Complejidad cognitiva o de la informaci√≥n. |
| **Exigencia Temporal** | Gran presi√≥n de tiempo o ritmo de trabajo r√°pido. | Problemas de plazos o cadencia de la tarea. |
| **Esfuerzo** | Alto grado de recursos invertidos para mantener el rendimiento. | El sujeto tuvo que **esforzarse mucho** para tener √©xito. |
| **Frustraci√≥n** | Alto nivel de sentirse irritado, estresado o descontento. | Dise√±o deficiente o interrupciones. |
| **Exigencia F√≠sica** | Alto requerimiento de esfuerzo muscular o movimiento. | Problemas ergon√≥micos. |
| **Rendimiento** (Invertida) | El sujeto siente que su **rendimiento fue deficiente** (Puntuaci√≥n bruta original fue baja). | Fracaso en el logro de objetivos, baja sensaci√≥n de eficacia. |

### üö® Recordatorio Clave: Inversi√≥n del Rendimiento

Para el c√°lculo final ponderado, la puntuaci√≥n de **Rendimiento (Performance)** se debe **invertir** para que una puntuaci√≥n baja en la subescala contribuya a una carga mental alta en el resultado final:

$$R'_{\text{OP}} = 100 - R_{\text{OP}}$$
